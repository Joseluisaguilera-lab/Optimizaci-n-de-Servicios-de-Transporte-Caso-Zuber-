```python
#Cargar el dataset para la prueba de hipótesis
trips_data = pd.read_csv('/datasets/project_sql_result_07.csv')
trips_data.head()
```

```python
#Informacion general del dataset
trips_data.info()
```

```python
#Convertir start_ts a formato datetime
trips_data['start_ts'] = pd.to_datetime(trips_data['start_ts'])

#Verificar que la conversion fue exitosa
print(trips_data['start_ts'].dtype)
print(trips_data['start_ts'].head())
```

```python
#Crear una nueva columna con el dia de la semana
trips_data['day_of_week'] = trips_data['start_ts'].dt.dayofweek

#Ver los primeros registros para verificar
print(trips_data[['start_ts', 'day_of_week']].head())
```

```python
# Verificar que todos los datos son de sábados
print("Días de la semana presentes:")
print(trips_data['day_of_week'].value_counts().sort_index())
print(f"\nTotal de viajes: {len(trips_data)}")
```

```python
#Ver los valores unicos en weather_conditions
trips_data['weather_conditions'].unique()
```

```python
#Calcular la duracion promedio por condicion climatica
duraciones_promedio = trips_data.groupby('weather_conditions')['duration_seconds'].mean()
print(duraciones_promedio)
```

```python
#Estadisticas descriptivas por condicion climatica
estadisticas = trips_data.groupby('weather_conditions')['duration_seconds'].describe()
print(estadisticas)
```

```python
#Verificar condiciones climaticas
print("Condiciones climaticas en sabados:")
print(trips_data['weather_conditions'].value_counts())
```

```python
#Grupo 1: Viajes con buen clima
good_weather = trips_data[trips_data['weather_conditions'] == 'Good']['duration_seconds']
good_weather
# Grupo 2: Viajes con mal clima  
bad_weather = trips_data[trips_data['weather_conditions'] == 'Bad']['duration_seconds']
bad_weather
```

```python
# Definir hipótesis
# H0: No hay diferencias en la duración promedio entre condiciones climáticas
# H1: La duración promedio es diferente entre buen y mal clima

alpha = 0.05
results = st.ttest_ind(bad_weather, good_weather, equal_var=False)

if results.pvalue < alpha:
    print(f'Rechazamos la hipótesis nula. Las duraciones promedio son diferentes entre condiciones climáticas.')
    print(f'p-value: {results.pvalue}')
else:
    print(f'No rechazamos la hipótesis nula.')
    print(f'p-value: {results.pvalue}')
```

Con respecto a los datos encontrados se lograron identificar varios puntos los cuales explican el por que los viajes duran mas en mal clima.

-Los conductores manejan con precaucion, las condiciones resbaladizas requieren velocidades menores, mayor riesgo de accidentes obliga a ser mas cuidadosos, todo esto se traduce en tiempos de viaje mas largos.
Tambien es importante mencionar la diferencia que se encontro: 40.5 minutos en mal clima vs 33.3 minutos en buen clima,¡una diferencia de mas de 7 minutos!

Bien hecho con el test de hipótesis, fue realizado de forma correcta. 

Es importante mencionar que el t-test asume que las dos muestras tienen varianzas iguales. Para verificar esa igualdad de varianzas podemos usar el **test de Levene**. Este nos dice si la variabilidad entre los grupos es estadísticamente diferente. En corto, el flujo sería así

1. Usamos `levene()` para verificar si las varianzas son iguales.
2. Dependiendo del resultado, usamos `ttest_ind()` con el argumento `equal_var`:
   - `equal_var=True` si Levene dice que las varianzas son iguales (p > 0.05).
   - `equal_var=False` si Levene dice que las varianzas son diferentes (p < 0.05).
